Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting styles
%matplotlib inline
sns.set(style='whitegrid')
```

Load the Data set 

```python
df = pd.read_csv('../data/WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.head()
```

Basic Dataset Overview

```python
# Check dataset shape and data types
print("Shape:", df.shape)
df.info()
```

```python
# View statistical summary
df.describe()
```

Null & Duplicate Check

```python
# Check for missing values
df.isnull().sum()
```

```python
# Drop duplicates
df.drop_duplicates(inplace=True)
```

Convert Object Columns to Categories

```python
# Convert object (text) columns to categorical for efficiency
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype('category')

df.dtypes.head(10)
```

Value Counts for Target Column

```python
# Target column distribution
df['Attrition'].value_counts(normalize=True)
```

Visualize Attrition Counts

```python
# Bar chart of Attrition
sns.countplot(data=df, x='Attrition')
plt.title('Employee Attrition Distribution')
```

```python
plt.figure(figsize=(12,5))
sns.countplot(data=df, x='JobRole', hue='Attrition')
plt.title('Attrition by Job Role')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df[df['Attrition'] == 'Yes'], x='Age', label='Yes')
sns.kdeplot(data=df[df['Attrition'] == 'No'], x='Age', label='No')
plt.title('Age Distribution by Attrition')
plt.legend()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome')
plt.title('Monthly Income by Attrition')
```

```python
# Encode categorical columns temporarily for correlation
df_encoded = df.copy()
for col in df_encoded.select_dtypes(include='category').columns:
    df_encoded[col] = df_encoded[col].cat.codes

# Correlation matrix
plt.figure(figsize=(16,12))
corr = df_encoded.corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
```
